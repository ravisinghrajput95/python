```python
server_name = ["web01", "web02"]
status_code = 404
package_names = ["pandas", "requests", "pip", "flask"]
cpu_usage = 60

if server_name:
    print(f"server names are:{server_name}")
    if server_name:
        print(f"number of servers are", len(server_name))
else:
    print(f"server names does not exists")

if status_code == 200:
    print(f"service is healthy and reachable")
elif status_code == 302:
    print(f"service is redirected to some other")
elif status_code == 404:
    print(f"resource not found")
elif status_code >= 500:
    print(f"server error (5xx)")
else:
    print(f"please provide valid status code")

if 'pandas' in package_names:
    print(f"package pandas exists in the package list")
    print(f"number of packages are : {len(package_names)}")
    if len(package_names) >=4:
        print(f"package count is as expected")
    else:
        print(f"package count should be >=4")
else:
    print(f"please provide non empty package names")

if 'sqlite' not in package_names:
    print(f"'sqlite' package does not exist")
    print(f"updating the list of package names to include sqlite package")
    package_names.append("sqlite")
    print(package_names)
    if 'sqlite' in package_names:
        print(f"package names are updated")
    else:
        print(f"package name are not updated")
else:
    print(f"sqlite is not required")

if cpu_usage >=90:
    print(f"'system' {server_name[0]}'s cpu_usage is : {cpu_usage}")
    print(f"server may be down or sluggish")
else:
    print(f"no need to worry, cpu usage is {cpu_usage}")
```
