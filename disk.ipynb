```python
server_name = "webserver-03"
cpu_cores = 4
memory_gb = 8.0
disk_total_gb = 500
disk_used_gb = 350

print(f"Server name is: {(server_name.upper())}")
print(f"Server has {(cpu_cores)} core CPUs and {(memory_gb)} GB of RAM ")
disk_usage = (disk_used_gb / disk_total_gb)*100
print(disk_usage)
formatted_disk_usage = "{:.2f}".format(disk_usage)
print(f"Disk Usage percentage of the server {(server_name)} is: {(formatted_disk_usage)}")


```
