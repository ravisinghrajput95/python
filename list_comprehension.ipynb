```python
numbers = [1,2,3,4,5]

squared_numbers = [ number**2  for number in numbers ]

print(squared_numbers)
```

```python
names = ["Ravi", "Veer", "Sanju", "Sangharsh"]

name_lowercase = [ name.lower() for name in names ]

print(name_lowercase)
```

```python
ages = [12, 15, 10, 18, 24]

new_ages = [age for age in ages if age > 18]

print(new_ages)
```

```python
words = ["abhay", "varun", "ritesh", "ravi", "mohit"]

word_vowel = [word for word in words if word.startswith(("a", "e", "i", "o", "u"))]

print(word_vowel)
```

```python
celsius = [29.5, 36, 40]

fahrenheit = [(temp * 9/5) + 32 for temp in celsius]

print(fahrenheit)
```

```python
sentences = ["The cat sat on the mat", "She drinks tea in the morning", "What is your name?", "Have you finished your homework?"]

first_word = [ sentence.split()[0] for sentence in sentences ]

print(first_word)
```

```python
usernames = ["Gaurav", "Pranit", "Amaal", "Tanya", "Farhana", "Ashnoor"]

new_list = [ username for username in usernames if len(username)> 6]

print(new_list)
```

```python
names = [("Gaurav", "Khanna"), ("Pranit", "More")]

full_names = [f"{first} {last}" for first, last in names]

print(full_names)
```

```python
product_price = [ 100, 200, 300, 400, 500 ]

discounted_price = [ price - (price * (20/100)) for price in product_price ]

print (discounted_price)
```

```python
numbers = [2, 6, 3, 4, 5 ]

new_numbers = [ number*3 for number in numbers if number%2!=0 ]

print(new_numbers)
```
