```python
attempts = [25, -3, 50, 7]
index = 0
num = attempts[index]

while num < 1 or num > 10:
    print("Invalid number:", num)
    index += 1
    num = attempts[index]

print("Valid number received:", num)
```

```python
numbers = [15, 30, 45, 50, 10, 25, 35, 40]  # sample list
total = 0
index = 0

while total < 200 and index < len(numbers):
    total += numbers[index]
    print("Added:", numbers[index], "| Total:", total)
    index += 1

print("Stopped — total reached:", total)
```
