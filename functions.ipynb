```python
def count_vowels(input_string):
    """
    Counts the number of vowels (a, e, i, o, u, case-insensitive) in a given string.
    """
    vowels = "aeiouAEIOU"
    vowel_count = 0
    for char in input_string:
        if char in vowels:
            vowel_count += 1
    return vowel_count

text = "Hello World"
count = count_vowels(text)
print(f"The number of vowels in '{text}' is: {count}")

text_2 = "Python Programming"
count_2 = count_vowels(text_2)
print(f"The number of vowels in '{text_2}' is: {count_2}")
```

```python
def calculate_discount(price, percent):
    discounted_price = price - price * (percent/100)
    return discounted_price

result = calculate_discount(400, 30)
print(result)
```

```python
def filter_even(numbers):
    even_numbers = []
    for number in numbers:
        if number%2==0:
            even_numbers.append(number)
    return even_numbers

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter_even(numbers)

print(even_numbers)
```

```python
def is_valid_password(password):
    if len(password)<8:
        return False

    has_number = False
    has_special_symbol = False
    special_symbols = {'@', '#', '$', '%'}

    for char in password:
        if char.isdigit():
            has_number = True
        if char in special_symbols:
            has_special_symbol = True

    return has_number and has_special_symbol

is_valid_password("Raviwdw1234")
```

```python
def get_down_servers(status_list):
    down_servers = []
    for server_info in status_list:
        if server_info.get('status') == "DOWN":
            down_servers.append(server_info)
    return down_servers

server_statuses = [
    {'server': 'Web-01', 'status': 'UP'},
    {'server': 'API-02', 'status': 'DOWN'},
    {'server': 'DB-01', 'status': 'UP'},
    {'server': 'Cache-03', 'status': 'DOWN'},
    {'server': 'Auth-04', 'status': 'UP'}
]

servers_down = get_down_servers(server_statuses)
print(servers_down)
```

```python
def filter_yaml(files):
    filtered_list = []
    for filename in files:
        # Check if the filename ends with .yaml or .yml, ignoring case
        if filename.lower().endswith(('.yaml', '.yml')):
            filtered_list.append(filename)
    return filtered_list

file_list = [
    "config.yaml",
    "data.yml",
    "image.png",
    "README.md",
    "CaseSensitive.YAML",
    "invalid.yamll"
]

yaml_files = filter_yaml(file_list)
print(f"Original list: {file_list}")
print(f"Filtered list: {yaml_files}")
```
