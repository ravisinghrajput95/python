```python
str1="Ravi Rajput"
str2="Sr DevOps Engineer"

print(f"Explore String functions on the given string")

"""
String concatenation
"""
print(f"Concat both the strings is: {(str1 + " " + "is" + " " + "a" + " " + str2)}")

# store the concat value in new variable

str_concat = str1 + " " + "is" + " " + "a" + " " + str2
print(str_concat)

# String functions in action

str_cap = str_concat.capitalize()
print(str_cap)

str_title = str_concat.title()
print(str_title)

str_lstrip = str_concat.lstrip()
print(str_lstrip)

str_rstrip = str_concat.rstrip()
print(str_rstrip)

str_lower = str_concat.islower()
print(str_lower)

str_startswith = str_concat.startswith("Ravi")
print(str_startswith)

str_ljust = str_concat.ljust(20)
print(str_ljust)

str_strip = str_concat.strip()
print(str_strip)
```
