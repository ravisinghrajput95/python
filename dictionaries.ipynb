```python
my_dict = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}

print(type(my_dict))

## Accessing elements
print(my_dict.keys())
print(my_dict.values())
print(my_dict.items())

my_dict_copy = my_dict.copy()
print(my_dict_copy)

my_dict_copy.clear()
print(my_dict_copy)

print(my_dict.fromkeys(my_dict))

removed = my_dict.pop('e')
print(f"Removed: {removed}")
print(my_dict)

update_dict = my_dict.update({"e": 5})
print(my_dict)

get_values = my_dict.values()
print(get_values)

dict_set_default = my_dict.setdefault("a")
print(dict_set_default)

## Membership test

print(f"'f' is in my_dict? {'f' in my_dict}")

print(f"'1' is in my_dict? {1 in my_dict.values()}")
```

```python
default_tags = {
    "Environment" : "Production",
    "Owner" : "DevOps team",
    "Cost Center" : "10000",
    "Tier" : "Frontend"
}

custom_tags = {
    "Cost Center" : "20000",
    "Owner" : "Apps team"
}

merged_tags = default_tags | custom_tags
print(merged_tags)

tags = {
    "Name" : "WebServer01",
    "Created by" : "DevOps"
}

tags["Name"] = "AppServer01"
tags["Created by"] = "Apps"

print(tags)
```

```python
## Declare the dictionary named as 'server_info'
server_info = {
    'id' : 'instance-01',
    'ip_address' : '172.10.11.01',
    'state' : 'started',
    'tags' : {
        'environment' : 'production',
        'owner' : 'engineering'
    }
}

## print server_info
print(server_info)

## Get the current server status
print(f"server status is : {server_info.get('state')}")

## Get instance type of the server
instance_type = server_info.get('instance_type', 't2.micro')
print(f"instance type is : {instance_type}")

## update server state to stopped
server_info.update({'state' : 'stopped'})
print(f"updated server status is : {server_info}")

## add a new tag to tags dictionary
server_info["tags"] ["region"] = "eu-central-1"

## get all the dict items
for item_tuple in server_info.items():
    key,value = item_tuple
    print(f" - {key}: {value}")
```
