```python
def server_health(server_name, region):
    if not server_name:
        print(f"server name is not provided")
        return
    if not region:
        print(f"region is not provided")
        return
    print(f"Checking health of {server_name} in {region}")

server_health("appserver01", "us-east-1")
```

```python
def register_user(email, password, country):
    if not email:
        print(f"email is not provided")
        return
    if not len(password) >=8:
        print(f"length of the password should be greater than or equal to 8")
        return
    if country not in ("US", "CA"):
        print(f"valid country is not provided")
        return
    print(f"user from {country} with {email} is registered with valid password{password}")

register_user("ravi@gmail.com", "DevOps@2025", "US")
```

```python
def api_request(url, api_key):
    if not url:
        print(f"url is not provided")
        return
    if not url.startswith("https"):
        print(f"enter valid url to proceed")
        return
    if not api_key:
        print(f"api_key is missing, cannot proceed")
        return
    print(f"Request sent")

api_request("http://api.com", "1aeffsfds")
```

```python
import os

def process_file(filename):
    if not filename:
        print(f"filename is missing, cannot proceed")
        return
    if not os.path.exists(filename):
        print(f"file is not present")
        return
    print(f"file name is processed")

process_file("file.txt")
```

```python
def database_connectivity(host, port, timeout):
    if not host:
        print("Host is not provided")
        return

    if not isinstance(port, int):
        print("Port should be an integer")
        return

    if timeout <= 0:
        print("Timeout should be greater than zero")
        return

    print("Connected")

database_connectivity("dbserver01", 5432, 1)
```

```python
def health_check(vm_id, region, status):
    if not vm_id:
        print(f"vm_id is not provided")
        return
    if not region:
        print(f"region is not provided")
        return
    if not status=="running":
        print(f"status is not running")
        return
    print(f"Health OK")

health_check("1asddsdf", "eu-central-1", "running")
```
