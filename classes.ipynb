```python
class person:
    """ Class to greet a person """
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, my name is {self.name}, My age is {self.age} years "

p1 = person("Alice", 29)
msg = p1.greet()
print(msg)
```

```python
class rectangle:
    """ find area and perimeter of a rectangle """
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        area_of_rectangle = self.width * self.height
        #return f"area of the rectangle is ", area_of_rectangle
        return area_of_rectangle

    def perimeter(self):
        perimeter_of_rectangle = 2*(self.width + self.height)
        return perimeter_of_rectangle
        

width_height = rectangle(5,6)
ar = width_height.area()
print(f"area of rectangle is {ar}")

pm = width_height.perimeter()
print(f"perimeter of rectangle is {pm}")
```

```python
class BankAccount:
    """ Bank account for deposit and withdrawal """
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("amount must be greater than 0")
        if amount > self.balance:
            raise ValueError("Insufficient balance")
        self.balance -= amount
        return self.balance


# Using the class
details = BankAccount("Alice", 1000)

deposit_amount = details.deposit(100)
print(f"Balance after deposit: {deposit_amount}")

withdraw_amount = details.withdraw(200)
print(f"Balance after withdrawal: {withdraw_amount}")
```

```python
class Car:
    def __init__(self, brand, model, speed=0):
        self.brand = brand
        self.model = model
        self.speed = speed

    def accelerate(self, amount=10):
        self.speed += amount
        return f"Car accelerated. Current speed: {self.speed} km/h"

    def brake(self, amount=10):
        if self.speed - amount < 0:
            self.speed = 0
        else:
            self.speed -= amount
        return f"Car slowed down. Current speed: {self.speed} km/h"


car1 = Car("Tata", "Nexon", 80)

print(car1.accelerate())
print(car1.brake())
print(car1.brake(50))
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def average(self):
        if len(self.marks) == 0:
            return 0
        return sum(self.marks) / len(self.marks)

    def passed(self):
        return self.average() > 40

s1 = Student("Ravi", [50, 60, 40])
s2 = Student("Aman", [20, 30, 25])

print(f"{s1.name} average: {s1.average()} | Passed: {s1.passed()}")
print(f"{s2.name} average: {s2.average()} | Passed: {s2.passed()}")
            
```

```python
class ShoppingCart:
    def __init__(self, items):
        self.items = items

    def add_item(self, item):
        self.items.append(item)
        return f"Added {item}"

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            return f"Removed {item}"
        return f"{item} not found"

    def total_items(self):
        return len(self.items)


cart = ShoppingCart(["Banana", "Soap", "Bread", "Eggs"])

print(cart.add_item("Pen"))
print(cart.remove_item("Soap"))
print("Total items:", cart.total_items())

```

```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def isgood(self):
        if not isinstance(self.rating, (int, float)) or not (0 <= self.rating <= 5):
            raise ValueError("Rating must be a number between 0 and 5")
        return "Good" if self.rating >= 4 else "Bad"


mv = Movie("Dhurandhar", 0.1)
print(mv.isgood())
```

```python
class Server:
    def __init__(self, name, region, status="inactive"):
        self.name = name
        self.region = region
        self.status = status

    def activate(self):
        self.status = "active"

    def deactivate(self):
        self.status = "inactive"

    def is_active(self):
        return self.status == "active"


s = Server("web01", "us-east-1")

s.activate()
print(s.is_active())

s.deactivate()
print(s.is_active())
```

```python
class Pod:
    def __init__(self, name, namespace, status="Pending"):
        self.name = name
        self.namespace = namespace
        self.status = status

    def restart(self):
        self.status = "Restarting"
        # simulate restart
        self.status = "Running"
        return f"{self.name} restarted successfully."

    def ready(self):
        return self.status == "Running"

p = Pod("auth-service", "dev", "Crashed")

print(p.restart())
print(p.ready()) 
```
