```python
statuses = ["DOWN", "DOWN", "MAINTENANCE", "BOOTING", "OK"]  # simulated checks
index = 0

status = statuses[index]
while status != "OK":
    print("Server status:", status)
    index += 1
    status = statuses[index]

print("Server is OK now!")
```

```python
logs = [
    "INFO: system working",
    "INFO: checking disk",
    "WARNING: memory high",
    "ERROR: disk failure",
    "INFO: recovered"
]

for line in logs:
    if "ERROR" in line:
        print("Error found:", line)
        break
else:
    print("No error found in logs")
```
