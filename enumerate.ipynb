```python
#Numbered Deployment Steps

steps = ["Pull image", "Start container", "Run migrations", "Verify health"]

for idx, step in enumerate(steps, 1):
    print(f"{idx}. {step}")
```

```python
# Monitoring Servers with Index

servers = ["web01", "web02", "db01"]

for idx, server in enumerate(servers):
    print(f" Checking server {idx}: {server}")
```

```python
# Log lines with line numbers

log_lines = ["Started", "Connected to DB", "Timeout"]
for idx, log in enumerate(log_lines, 1):
    print(f"Line {idx}: {log}")
```

```python
# Kubernetes Pods With Serial Numbers

pods = ["auth", "frontend", "backend"]

for idx, pod in enumerate(pods, 1):
    print(f"{idx}){pod}")
```

```python
# Identify the first failing test case

results = ["PASS", "PASS", "FAIL", "PASS"]

for idx, result in enumerate(results):
    if result=="FAIL":
        print(f"First Failure at index {idx} ")
        break
```
