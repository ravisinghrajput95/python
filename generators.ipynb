```python
def count_up_to(n):
    for num in range(1, n + 1):
        yield num
        
for number in count_up_to(7):
    print(number)
    
```

```python
def evens(n):
    for num in range(1, n+1):
        if num % 2 ==0:
            yield num

for number in evens(30):
    print(number)
```

```python
def list_items(items):
    for item in items:
        yield item

items = ["Pen", "Paper", "Clothes", "Scissors"]
for it in list_items(items):
    print(it)
```

```python
def squares(numbers):
    for number in numbers:
        square = number * number
        yield square

list_numbers = [1, 2, 3, 4, 5, 6]
for num in squares(list_numbers):
    print(num)
```

```python
def filename(file):
    with open(file) as f:
        for line in f:
            yield line

for line in filename("file.txt"):
    print(line.strip())
```

```python
def read_logs(logs, keyword="error"):
    with open(logs) as f:
        #word = "error"
        for line in f:
            if keyword.lower() in line.lower():
                yield line.rstrip("\n")

for line in read_logs("logs.txt", keyword="error"):
    print(line)
```

```python
def chunk_data(data, size):
    for i in range(0, len(data), size):
        yield data[i:i + size]

large_list = list(range(1, 21))

for chunk in chunk_data(large_list, 5):
    print(chunk)
```
