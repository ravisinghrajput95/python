```python
numbers = [1, 2, 3, 4]
result = list(map(lambda x: x * 10, numbers))
print(result)
```

```python
names = ["ravi", "prashant"]

result = list(filter(lambda x: len(x)>4, names))
print(result)
```

```python
servers = [
    {"name": "web01", "disk": 70},
    {"name": "db01", "disk": 90},
    {"name": "cache01", "disk": 40}
]
sorted_servers = sorted(servers, key=lambda s: s["disk"])
print(sorted_servers)
```

```python
services = [
    {"name": "auth", "status": "Running"},
    {"name": "db", "status": "Running"},
    {"name": "cache", "status": "Running"}
]
all_running = all(map(lambda s: s["status"] == "Running", services))
print(all_running)
```

```python
names = ["ravi", "ashish", "aman", "kartik"]
upper_names = list(map(lambda n: n.upper(), names))
print(upper_names)
```
