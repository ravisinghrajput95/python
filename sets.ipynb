```python
required_packages = {"nginx", "nginx", "apache2", "tree", "pip", "requests"}

required_packages.add("paramiko")
print(required_packages)

required_packages.discard("pip")
print(required_packages)

installed_packages = {"nginx", "tree"}

print("requests" in required_packages)
print("ansible" in required_packages)

print("common_packages:",required_packages.intersection(installed_packages))
print("missing_packages:", required_packages.difference(installed_packages))
print("extra_packages:", installed_packages.union(required_packages)) 
```
