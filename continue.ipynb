```python
numbers = [2, 5, 9, 0, 13]

for number in numbers:
    if number % 2 == 0:
        continue
    print(number)
        
```

```python
filenames = ["file.txt", "values.yaml", "main.tf", "sample.tmp"]

for file in filenames:
    if file.endswith(".tmp"):
        continue
    print(file)
```

```python
students = ["Ravi", "Ashish", "Pallavi", "Deepali"]

for student in students:
    if student.startswith("A"):
        continue
    print(student)
```

```python
## Skip "Timeout" messages and print all other responses.

statuses = ["DOWN", "DOWN", "MAINTENANCE", "BOOTING", "OK", "Timeout"]
index = 0

while index < len(statuses):
    status = statuses[index]
    if status == "Timeout":
        index += 1
        continue
    print("Server status:", status)
    index += 1
```

```python
processes = ["running", "completed", "pending", "failed", "completed", "stopped"]

for process in processes:
    if process == "completed":
        continue
    print("Processing:", process)
```
