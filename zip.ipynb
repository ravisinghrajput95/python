```python
# Pair usernames with emails

users = ["alice", "bob", "charlie"]
emails = ["alice@mail.com", "bob@mail.com", "charlie@mail.com"]

for user, email in zip(users, emails):
    print(f"{user} -> {email}")
```

```python
# Combine server names and their statuses

servers = ["web01", "web02", "db01"]
status  = ["UP", "UP", "DOWN"]

for server, stat in zip(servers, status):
    print(f"{server} is {stat}")
```

```python
# Kubernetes pods + namespaces

pods       = ["auth", "frontend", "db"]
namespaces = ["dev", "prod", "prod"]

for pod, ns in zip(pods, namespaces):
    print(f"{pod} in {ns}")
```

```python
# Add corresponding CPU + Memory usage

cpu = [30, 50, 70]
memory = [512, 1024, 2048]

for c, m in zip(cpu, memory):
    print(f"CPU: {c}, Memory: {m}")
```

```python
# Merge two lists into a dictionary

keys = ["name", "env", "port"]
values = ["backend", "prod", 8080]

merged_dict = dict(zip(keys, values))
print(merged_dict)
```
